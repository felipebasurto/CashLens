```python
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

print(device)
```

```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO('cashlens.pt')

# Open the webcam
cap = cv2.VideoCapture(1) # 0 is for first webcam, use 1 for second webcam, etc.

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)
        
        total_amount = 0
        names=results[0].names

        # Visualize the results on the frame
        annotated_frame = results[0].plot() # render returns a list of images
                
        for box in results[0].boxes:
            id = box.cls.item()
            coin_value = names[id]
            total_amount += int(coin_value)

        #Display the total amount in euros, because right now is in cents
        cv2.putText(annotated_frame, str(total_amount/100) + "Euro", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
        
        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)
        
        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
